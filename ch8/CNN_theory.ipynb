**합성곱**<br>
각 특성에 가중치를 곱해서 출력을 만듦<br>
$a_1*w_1 + a_2*w_2 + ... + a_n*w_n + b$<br>
단, 입력 전체에 가중치를 적용하지 않고 일부에 가중치를 곱하며, 이것을 반복<br>
$a_1*w_1 + a_2*w_2 + a_3*w_3 = output_1$<br>
$a_2*w_2 + a_3*w_3 + a_4*w_4 = output_2$<br>
...<br>
곱해지는 가중치의 모음을 **필터(커널)**이라 부름 (위 그림 참고)<br>
합성곱 특성상 2차원 입력에 적용하고 2차원 형태를 유지할 수 있으며, 따라서 이미지 처리 분야에 뛰어남<br>


image.png

서로 다른 가중치를 가진 필터를 여러 개 사용할 수 있으며, 여기서 여러 개의 특성 맵을 만들어 쌓을 수 있음<br>
(**특성 맵**: 합성곱 계산을 통해 얻은 출력)<br>
2차원의 특성 맵을 여러 개 쌓으면 3차원 배열이 됨

**합성곱 신경망**<br>
합성곱 층을 1개 이상 사용한 인공 신경망

image.png

**패딩**<br>
커널이 출력을 더 많이 만들기 위해 입력 배열 주변을 가상의 원소로 채우는 것 (보통 0으로 채움)<br>
 * 세임 패딩: 입력과 특성 맵의 크기를 동일하게 만들기 위해 입력 주위를 0으로 패딩
 * 밸리드 패딩: 패딩 없이 순수한 입력 배열에서만 합성곱을 하여 특성 맵을 만드는 경우


image.png

패딩을 하지 않을 경우, 모서리 부분의 픽셀이 합성곱에 덜 참여<br>
-> 모서리의 중요한 정보가 누락될 수 있음

image.png

**스트라이드**<br>
필터가 몇 칸씩 이동하는지를 나타냄<br>
스트라이드가 클수록 특성 맵의 크기는 작아짐<br>
(대부분 스트라이드 값으로 1을 사용함)

image.png

**풀링**<br>
합성곱 층에서 만든 특성 맵의 가로세로 크기를 줄임
 * 최대 풀링: 각 영역에서 최댓값을 고름
 * 평균 풀링: 각 영역에서 평균값을 고름

image.png

**합성곱 신경망의 전체 구조**<br>
1. 합성곱 층 -> 세임 패딩 적용
2. 풀링층 -> 보통 (2,2) 풀링을 사용해 절반으로 줄임
3. 밀집층 -> 3차원 배열을 1차원으로 펼치고 분류 수행

image.png

컬러 이미지 - 입력이 3차원 (높이 x 너비 x 3)<br>
위의 숫자 3: R,G,B 값<br>
->커널도 3차원<br>
입력의 깊이와 필터의 깊이는 같음<br>
훈련이 진행될수록 너비와 높이는 줄어들고 깊이는 점점 깊어짐
